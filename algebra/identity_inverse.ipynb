```julia
using LinearAlgebra

# Identity matrix

I3 = I(3)
println("Identity 3x3:")
display(Matrix(I3))

x = [2.0, 5.0, -1.0]
println("\nI* x = x: ", I3 * x == x)
```

```julia
# Inverse matrix
A = [4.0 7.0; 2.0 6.0]
A_inv = inv(A)

println("\nMatrix A:")
display(A)
print(A)

println("\nInverse A⁻¹:")
display(A_inv)
print(A_inv)

println("\nA * A⁻¹ ≈ I: ", A * A_inv ≈ I(2))
```

```julia
# Solving linear system Ax = b

A = [3.0 1.0; 1.0 2.0]
b = [9.0, 8.0]

x_direct = A \ b
x_inverse = inv(A) * b

println("\nSolving Ax = b:")
println("Direct solve (A\\b): ", x_direct)
println("Via inverse: ", x_inverse)
println("Residual ∥Ax - b∥: ", norm(A * x_direct -b))
```

```julia
# ML example: Linear regression Closed Base
# ordinary least squares in closed form using normal equations

X = [1.0 1.0; 1.0 2.0; 1.0 3.0; 1.0 4.0]
y = [2.1, 3.9, 6.1, 7.9]

w = inv(X' * X) * X' * y
println("\nLinear regression weights (closed form):")
println("w = ", w)
println("Predictions: ", X * w)

```

```julia
# Numerical stability

A_nice = [1.0 0.0; 0.0 1.0]
A_bad = [1.0 1.0; 1.0 1.0 + 1e-10]

println("\nCondition numbers:")
println("Well-conditioned: ", cond(A_nice))
println("Ill-conditioned: ", cond(A_bad))
```
