```julia
using Statistics
# Regression: Minimise E[(y - ŷ)²]
# This is the expectation of squared error


# Samples data
X = randn(100, 1)
y_true = 2 .* X .+ 1 .+ 0.1*randn(100, 1) # y = 2x + 1 + noise

# Model: ŷ = w*x + b
w, b = 0.0, 0.0

# Loss = mean squared error
function loss(w, b, X, y)
    predictions = w .* X .+ b
    errors = predictions .- y
    return mean(errors.^2)
end

# Gradient of loss
function grad_loss(w, b, X, y)
    n = length(y)
    predictions = w .* X .+ b
    errors = predictions .- y

    dL_dw = (2/n) * sum(errors .* X) # Partial derivative wrt w
    dL_db = (2/n) * sum(errors)

    return dL_dw, dL_db
end

current_loss = loss(w, b, X, y_true)
dw, db = grad_loss(w, b, X, y_true)

println("Current loss: $current_loss")
println("Gradient: [dL/dw=$dw, dL/db=$db]")
println("These tell us how to change w,b to reduce loss")
```
