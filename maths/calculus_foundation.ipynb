```julia
using Pkg

Pkg.activate(mktempdir())
Pkg.update()

Pkg.add([
    "Plots"
])
```

```julia
# Visualise: derivative is the slope at a point
using Plots

f(x) = x^2
df_dx(x) = 2x # derivative

x = 2.0
tangent(t) = f(x) + df_dx(x) * (t - x) # Tangent line at x=2

plot(f, -1, 4, label="f(x) = x²", lw=2)
plot!(tangent, 0, 4, label="Tangent (slope = 2x)", lw=2)
scatter!([x], [f(x)], label="Point x=2", ms=4)
```

```julia
# If z = f(g(x)), then dz/dx = (df/dg) × (dg/dx)

# Example: z = (x²)³
# Inner: g(x) = x², dg/dx = 2x
# Outer: f(g) = g³, df/df = 3g²

x = 2.0
# Method 1: Direct
z_direct = (x^2)^3

# Method 2: Chain rule
g = x^2
dg_dx = 2x
df_dg = 3g^2
dz_dx = df_dg * dg_dx

println("z = $z_direct")
println("dz/dx via chain rule = $dz_dx")

# Verify with finite differences
ϵ = 1e-7
approx_deriv = (((x + ϵ)^2)^3 - z_direct) / ϵ
println("Numerical check: $approx_deriv")
```

```julia
# For f(x, y) = x² + y², gradient is [∂f/∂x, ∂f/∂y]

f(x, y) = x^2 + y^2
grad_f(x, y) = [2x, 2y]

x, y = 1.0, 2.0
g = grad_f(x, y)

println("Gradient at ($x, $y): $g")
println("This points uphill - steepest increase direction")
println("Negative gradient: $(-g) points downhill")
```
