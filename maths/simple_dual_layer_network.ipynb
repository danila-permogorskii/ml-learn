```julia
using Pkg;

Pkg.activate(mktempdir())
Pkg.update()
```

```julia
# Simple 2-layer network
function forward(x, W1, W2)
    h = max.(0, W1 * x)  # ReLU activation
    y = W2 * h
    return y, h  # Save h for backward pass
end

function backward(x, y_pred, y_true, h, W1, W2)
    # Backward from output
    dL_dy = 2(y_pred .- y_true)
    
    # Gradient through W2
    dL_dW2 = dL_dy * h'
    
    # Gradient flows back to h
    dL_dh = W2' * dL_dy
    
    # Through ReLU (derivative is 0 or 1)
    dL_dh .*= (h .> 0)
    
    # Gradient through W1
    dL_dW1 = dL_dh * x'
    
    return dL_dW1, dL_dW2
end

# Test
W1 = randn(3, 2)
W2 = randn(1, 3)
x = randn(2)
y_true = 1.0

y_pred, h = forward(x, W1, W2)
dW1, dW2 = backward(x, y_pred, y_true, h, W1, W2)

println("Output: $y_pred, Target: $y_true")
println("Gradient shapes: W1=$(size(dW1)), W2=$(size(dW2))")
```
